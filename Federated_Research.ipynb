```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import numpy as np
import joblib
```

```python
from google.colab import files

# Load the dataset
df = pd.read_csv('heart.csv')

# Display the first few rows of the dataset
df.head()
```

```python
# Drop the output column
output_column_name = 'target'

# Divide the dataset based on 'sex'
client1_df = df[df['sex'] == 0].copy()  # Males
client2_df = df[df['sex'] == 1].copy()  # Females

# Display the shapes of the datasets to verify
print("Client 1 dataset shape:", client1_df.shape)
print("Client 2 dataset shape:", client2_df.shape)

client1_df

```

Client 1

```python
X_client1 = client1_df.drop("target", axis=1)  # Drop the "output" column to create features
y_client1 = client1_df["target"]


X_train_client1, X_test_client1, y_train_client1, y_test_client1 = train_test_split(X_client1, y_client1, test_size=0.2, random_state=42)

# Initialize logistic regression model
model_client1 = LogisticRegression(max_iter=1000)

# Train the model
model_client1.fit(X_train_client1, y_train_client1)

# Predict on test set
y_pred_client1 = model_client1.predict(X_test_client1)

# Evaluate accuracy
accuracy_client1 = accuracy_score(y_test_client1, y_pred_client1)
print("Accuracy for client 1 dataset on: ", accuracy_client1)


# Get the model weights (coefficients and intercept)
model_weights = {
    'coefficients': model_client1.coef_,
    'intercept': model_client1.intercept_,
}

# Define the path to save the model weights
model1_weights_path = '/content/model_client1_weights.pkl'

# Save the model weights
joblib.dump(model_weights, model1_weights_path)


# Extract features and target from df
X_df = df.drop("target", axis=1)
y_df = df["target"]

# Predict on df
y_pred_df = model_client1.predict(X_df)

# Evaluate accuracy
accuracy_df = accuracy_score(y_df, y_pred_df)
print("Accuracy on Compelete Dataset:", accuracy_df)
```

Client 2

```python
X_client2 = client2_df.drop("target", axis=1)  # Drop the "output" column to create features
y_client2 = client2_df["target"]


X_train_client2, X_test_client2, y_train_client2, y_test_client2 = train_test_split(X_client2, y_client2, test_size=0.2, random_state=42)

# Initialize logistic regression model
model_client2 = LogisticRegression(max_iter=1000)

# Train the model
model_client2.fit(X_train_client2, y_train_client2)

# Predict on test set
y_pred_client2 = model_client2.predict(X_test_client2)

# Evaluate accuracy
accuracy_client2 = accuracy_score(y_test_client2, y_pred_client2)
print("Accuracy for client 2 dataset:", accuracy_client2)


# Get the model weights (coefficients and intercept)
model_weights = {
    'coefficients': model_client2.coef_,
    'intercept': model_client2.intercept_,
}

# Define the path to save the model weights
model2_weights_path = '/content/model_client2_weights.pkl'

# Save the model weights
joblib.dump(model_weights, model2_weights_path)


# Extract features and target from df
X2_df = df.drop("target", axis=1)
y2_df = df["target"]

# Predict on df
y_pred_df = model_client2.predict(X2_df)

# Evaluate accuracy
accuracy_df = accuracy_score(y2_df, y_pred_df)
print("Accuracy on Compelete Dataset:", accuracy_df)
```

```python
# Path to the aggregated model weights file
aggregated_weights_path = '/content/aggregated_model_weights.pkl'

# Load the aggregated model weights
aggregated_weights = joblib.load(aggregated_weights_path)

model = joblib.load(aggregated_weights_path)

model
```

```python
import numpy as np
from sklearn.metrics import accuracy_score

coefficients = model['coefficients']
intercept = model['intercept']

# Separate the features and the target variable from the DataFrame
XWhole_df = df.drop("target", axis=1)
yWhole_df = df["target"]

# Convert the features DataFrame to a numpy array
input_data = XWhole_df.values

# Ensure input_data has the correct shape (number of features should match the coefficients)
assert input_data.shape[1] == coefficients.shape[1], "Mismatch in the number of features"

# Compute the logits (linear combination of input data and coefficients, plus intercept)
logits = np.dot(input_data, coefficients.T) + intercept

# Apply the sigmoid function to compute probabilities
probabilities = 1 / (1 + np.exp(-logits))

# Convert probabilities to class predictions (0 or 1) using a threshold of 0.5
class_predictions = (probabilities >= 0.5).astype(int)

# Calculate accuracy
accuracy = accuracy_score(yWhole_df, class_predictions)

# Print the overall probabilities, class predictions, and accuracy
# print("Overall Probability Mean:", np.mean(probabilities))
# print("Overall Class Prediction Mean:", np.mean(class_predictions))
print("Global Model Accuracy on complete dataset:", accuracy)
```
